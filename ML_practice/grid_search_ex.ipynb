```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../datasets/50_Startups.csv')
df.head()
```

```python
X = df[['R&D Spend']]
y = df['Profit']
```

```python
trainx,testx,trainy,testy = train_test_split(X,y,test_size=.2, random_state = 0)
```

```python
model = DecisionTreeRegressor()
model.fit(trainx,trainy)
```

```python
 model.score(testx,testy)*100
```

```python
# applying grid search
DecisionTreeRegressor?
```

```python
grid_params = {'criterion':['mse','friedman_mse','mae'],'splitter':['best','random'],'max_depth':[10,15,20],'max_features':["auto", "sqrt", "log2"],'random_state':[0]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(model,grid_params,cv=3,n_jobs=-1)
```

```python
grid.fit(X,y)
```

```python
grid.cv_results_
```

```python
grid.cv_results_.get('rank_test_score')
```

```python
grid.best_estimator_
```

```python
final_model = DecisionTreeRegressor(max_depth=10, max_features='auto', random_state=0)
final_model.fit(trainx,trainy)
final_model.score(testx,testy)*100
```
