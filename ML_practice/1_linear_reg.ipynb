```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd .read_csv('canada_per_capita_income.csv')
df.head()
```

```python
x = df[['year']]
y = df['per capita income (US$)']
```

```python
plt.scatter(x,y,c='red',marker='+')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.plot()
```

```python
trainx, testx, trainy,testy = train_test_split(x,y,test_size=.3,random_state=0)
```

```python
regr = LinearRegression()
regr.fit(trainx,trainy)
score = regr.score(testx,testy)*100
score
```

```python
ypred = regr.predict(trainx)
```

```python
regr.coef_
```

```python
regr.intercept_
```

```python
plt.scatter(trainx,trainy,c='red',marker="+",label='original')
plt.plot(trainx,ypred,linestyle='solid',label='prediction')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.legend()
plt.text(1970,45000,f'R2 = {score}')
plt.show()
```

```python
# prediction canada's per capita income in year 2020
regr.predict([[2020]])
```
